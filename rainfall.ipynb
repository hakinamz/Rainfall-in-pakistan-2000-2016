```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import lightgbm as lgb
import plotly.express as px
```

```python
rainfall=pd.read_csv("Rainfall_1901_2016_PAK.csv")
```

```python
#Basic Analysis
rainfall.head()
```

```python
rainfall.shape
```

```python
rainfall.info()
```

```python
#Renaming columns
rainfall.rename(columns ={' Year':'year'}, inplace=True)
rainfall.rename(columns ={"Month":'month'}, inplace=True)
rainfall.rename(columns ={"Rainfall - (MM)":'rainfall-mm'}, inplace=True)
```

```python
rainfall.head()
```

```python
#Rainfall(mm) in January from year 2000 to 2016
plt.figure(figsize=(10,6))
sns.barplot(x=rainfall["year"].loc[(rainfall.year >=2000)].loc[(rainfall.month =="January")],y=rainfall["rainfall-mm"]).set_title("Rainfall(mm) in January from year 2000 to 2016")
```
